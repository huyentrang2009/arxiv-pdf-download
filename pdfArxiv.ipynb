```python
from bs4 import BeautifulSoup 
import requests
import os
```

```python
def crawl_pdf():
    domain = "https://arxiv.org/list/cs.AI/recent"
    response = requests.get(domain)
    soup = BeautifulSoup(response.content, "html.parser")
    titles = soup.find_all("div", class_="list-title mathjax")

    return [t.get_text() for t in titles]

crawl_pdf()
```

```python
# Set the base URL and the number of PDFs to download
base_url = "https://arxiv.org/list/cs.AI/pastweek?skip={}&show=25"
num_pdfs = 200
save_dir = "/Users/yoohyra/Desktop/arxivorg/pdfs"

# Initialize the skip parameter
skip = 0

# Loop until we have downloaded the desired number of PDFs
while skip < num_pdfs:
    # Construct the URL for the current batch of PDFs
    url = base_url.format(skip)

    # Send an HTTP request to the webpage
    response = requests.get(url)

    # Parse the HTML using Beautiful Soup
    soup = BeautifulSoup(response.content, 'html.parser')

    # Extract the PDF links
    pdf_links = []
    for a in soup.find_all('a', {"title": "Download PDF"}):
        pdf_links.append(a['href'])

    # Download the PDF files
    for i, link in enumerate(pdf_links):
        pdf_url = f"https://arxiv.org{link}"
        response = requests.get(pdf_url, stream=True)
        if response.status_code == 200:
            with open(os.path.join(save_dir,f"pdf_{skip+i+1}.pdf"), 'wb') as f:
                for chunk in response.iter_content(1024):
                    f.write(chunk)
            print(f"Downloaded PDF {skip+i+1} to {save_dir}")
        else:
            print(f"Error downloading PDF {skip+i+1}")

    # Increment the skip parameter
    skip += 25
```
